5. Overlapping Zoomed Regions
-----------------------------

::: {#menu4}
[[Lesson](/documentation/tutorials/scaling/overlaping_zooms/lesson){.clean}]{.active}
[Images](/documentation/tutorials/scaling/overlaping_zooms/images){.normal}

:::

When overlapping zoom regions are defined, the zoom level at an ideogram
position is taken to be the highest (in absolute terms) zoom of any
overlapping regions.

For example, if you define the following regions

```ini
100-200Mb - 2x
150-180Mb - 3x
160-170Mb - 5x
```


then the net effect will be as if you defined

```ini
100-150Mb - 2x
150-160Mb - 3x
160-170Mb - 5x
170-180Mb - 3x
180-200Mb - 2x
```


The zoom precedence is defined by the absolute zoom level, which is
taken to be max(scale,1/scale) For example, a scale of 0.1 (absolute
zoom = 1/0.1 = 10x) overrides any zoom setting with an absolute zoom
level smaller than 10x.
