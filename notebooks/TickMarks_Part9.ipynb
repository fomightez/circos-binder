A tick set can be automatically drawn at multiple radial positions if
you define more than one radius for the set.

```ini
<tick>
# drawn at one radius
radius = dims(ideogram,radius_outer)
...
</tick>
```


```ini
<tick>
# drawn at two radii
radius = dims(ideogram,radius_outer)
radius = 0.8r
...
</tick>
```


```ini
<tick>
# drawn at three radii
radius = dims(ideogram,radius_outer)
radius = 0.8r
radius = 0.6r
...
</tick>
```


Tick rings are useful when your image has many tracks and you\'d like to
have multiple instances of tick marks.
